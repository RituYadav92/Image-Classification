```python
import torch
import numpy as np
import torch.nn as nn
import torch.backends.cudnn as cudnn
import matplotlib.pyplot as plt
import torchvision.models as models
from torch.autograd import Variable
from PIL import Image
import glob
import os
import pandas as pd
from transform import transform_training, transform_testing
from dataset import dataset
from train_test import start_train_test, test_only, test_sumission
```

```python
use_cuda = torch.cuda.is_available()
```

```python
trainloader, testloader, outputs, inputs = dataset('dog-breed')
print ('Output classes: {}\nInput channels: {}'.format(outputs, inputs)) 
```

```python
##ALEXNET
net1 = models.alexnet(pretrained=True)
for param in net1.parameters():
    param.requires_grad = False

net1.classifier._modules['6'] = nn.Linear(4096, outputs)
print(net1)
```

```python
#'''
##RESNET

net1 = models.resnet101(pretrained=True)
for param in net1.parameters():
    param.requires_grad = False

num_ftrs = net1.fc.in_features
net1.fc = nn.Linear(num_ftrs,120)
#net.add(Dense(NUM_CLASSES, activation='softmax'))
print(net1)

#'''
```

```python
if use_cuda:
    net1.cuda()
    print(torch.cuda.device_count())
    net = torch.nn.DataParallel(net1, device_ids=range(torch.cuda.device_count()))
    cudnn.benchmark = True
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
train_loss, test_loss = start_train_test(net1, trainloader, testloader, criterion) ##train,validate and save the model
```

```python
labels = []
image_id = []
count=0
prediction=[]

for filename in glob.glob('./Breed/Test2/test/*.jpg'): 
    
    image_name=(filename.split('.', -1)[1]).split('/',-1)[-1]
    image_id.append(image_name)
    
    im=Image.open(filename)
    
    im_transform=transform_testing()
    image=im_transform(im)
        
    image.unsqueeze_(0)
    out=net(image)
    
    m=nn.Softmax() ##prediction probability
    x = m(out)
    b = x.to(torch.device("cpu"))
    pred_array=b.detach().numpy()    
    prediction.append(pred_array)
    
    _, preds = torch.max(out, 1)
    count=count+1
```

```python
print(np.shape(prediction))
print(np.size(prediction))
print(type(prediction))
prob_prediction = np.reshape(prediction,(count, 120))
```

```python

label_list=os.listdir("./Breed/dataset1/train")
label_list_sort=sorted(label_list)
sol=pd.DataFrame(prob_prediction,columns=label_list_sort)
sol.insert(0,"Id",image_id)
np.shape(sol)
sol.to_csv("dogsBreed_res.csv", index = False)
```

```python
from mpl_toolkits.axes_grid1 import ImageGrid
def imshow(axis, inp):
    """Denormalize and show"""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    axis.imshow(inp)
```

```python
img, label = next(iter(trainloader))
img = img[100:104]
label = label[100:104]
print(img.size(), label.size())
fig = plt.figure(1, figsize=(16, 4))
grid = ImageGrid(fig, 111, nrows_ncols=(1, 4), axes_pad=0.05)    
for i in range(img.size()[0]):
    ax = grid[i]
    imshow(ax, img[i])
```
